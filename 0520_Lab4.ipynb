# Task4

```python
# while
print("Numbers < 10 (while):")
i = 0 #counter
while (i<10):
    print(i, end=" ") # print i in one line
    i += 1 # the counter increased "by hands"
print("\n")

# for
print("Numbers < 10 (for):")
for i in range(0,10): # the counter increased automatically
    print(i, end=" ")
else:
    print("\nThe last number is 10\n")

# break
sum = 0
for i in range(0,100):
    if i > 10:
        print("\nWe reached the end, final sum: ", sum)
        break
    sum += i

# continue
i = 0
while i<=15:
    if i % 3 == 0:
        i += 1
        continue #go to the next iteration
    print(i, end=" ")
    i += 1
print("\n")
```

# Task8

```python
a = [1,2,3,4,5]
print("a[0]: ", a[0])
print("List a[0:5]: ", a[0:5])
print("List a[:]: ", a[:])
print("List a: ", a)
print("List a[1:5]: ", a[1:])
print("List a[0:4]: ", a[:4])
print("Length of list a: ", len(a))

print("\nList a (by index):")
# getting list items in a loop (1)
for i in range(0,len(a)):
    print(a[i], end=" ")

print("\nList a (by elements):")
# getting list items in a loop (2)
for elem in a:
    print(elem, end=" ")
print("\n")

b = []
b.append(20) # adding an element to the end
b.extend(a) # adding list items a to b
print("List b (extended): ",b)

b.insert(3,5) # add an item to a position
print("List b (insert element): ",b)

b.remove(5) # delete the first element equal to 5
print("List b (remove element): ",b)

c1 = b.pop() # delete and return the last item of b
print("List b (pop last element): ",b)
print("c1: ",c1)

c2 = b.pop(3) # delete and return element from position
print("List b (pop 3rd element): ",b)
print("c2: ",c2)

bCopy = b.copy() # create a copy of the list
print("List b: ",b)
print("List bCopy: ",bCopy)

b.reverse() # change the ale. in places from the end to the beginning
print("List b (reversed): ",b)

b.sort(reverse=True) # sorting list items
print("List b (sorted): ",b)

b.clear(); # clearing the list
print("List b (cleared): ",b)
print("\n")

# list comparison
a1 = [1,2,4]
a2 = [1,2,3]

print("a1 == a2: ", a1 == a2)
a1.append(-1)
print("a1 == a2: ", a1 == a2)
print("a1 > a2: ", a1 > a2)
print("a1 < a2: ", a1 < a2)
print("\n\n")
```

# Task12

```python
bin_num = input("Enter a binary number: ")
bin_num_inverted = ""

for i in bin_num:
    if i == '0':
        bin_num_inverted += '1'
    elif i == '1':
        bin_num_inverted += '0'

print("the original num is : ", bin_num)
print("the inverted str is : ",bin_num_inverted)
```
