# Task1

```python
# creating a tuple
a1 = tuple()
a2 = 1, 2, 3, "abc"
a3 = (1, 2, 3, "abc")
print("Tuple a1 = ", a1)
print("Tuple a2 = ", a2)
print("Tuple a3 = ", a3)

# creating a tuple from other data structures
l = [1, 2, 3, "abc"] # from the list
a4 = tuple(l)
print("Tuple a4 from list l = ", a4)
a5 = tuple("Hello, World!") # from string
print("Tuple a5 from string = ", a5)

# combining of tuples
a6 = a2, a3
print("Tuple a6 formed by a2 and a3 = ", a6)

# union of tuples
a7 = a2 + a3
print("Tuple a7 by combining a2 and a3 = ", a7)

# accessing elements of tuples
print("a6[0]: ", a6[0])
print("a6[0][3]: ", a6[0][3])
# a6[0][3] = "cba"
print("\n")
```

# Task7

```python
print("Check 'bear' in b2 = ", "bear" in b2)
b4 = set("123456135")
b5 = set("12367")
print("Set b4: {0}, \nSet b5: {1}".format(b4,b5))
print("b4 - b5: ", b4 - b5) # elements in the first set, but not in the second
print("b4 difference b5 (b4-b5): ", b4.difference(b5))
print("b4 | b5: ", b4 | b5) # elements that are at least in one set
print("b4 union b5 (b4 | b5): ", b4.union(b5))
print("b4 & b5: ", b4 & b5) # elements in both sets

print("b4 intersection b5 (b4&b5): ", b4.intersection(b5))
print("b4 ^ b5: ", b4 ^ b5) # presence in only one of the sets

# checking for the intersection of sets
print("b4 and b5 are disjoint: ", b4.isdisjoint(b5))
b4.update(b5) # add elements of another set
print("add b5 to b4: ", b4)
b4.add("abc") # add element
print("add 'abc' to b4: ", b4)
b4.remove("5") # delete element
print("remove element '5' from b4: ", b4)
b4.clear() # clear set
print("clear b4: ", b4)
print("\n ")
```

```python
# create new sets
set1 = set("qetuwrt")
set2 = set("asfrewgq")

print("Set set1: {0}, \nSet set2: {1}".format(set1,set2))
# difference
print("set1 difference set2 (set1-set2): ", set1.difference(set2))
# union
print("set1 union set2 (set1 | set2): ", set1.union(set2))
# intersection
print("set1 intersection set2 (set1&set2): ", set1.intersection(set2))
# symmetric difference
print("set1 symmetric difference set2: ",set1.symmetric_difference(set2))
# add elementry by update
set1.update(set2)
print("set1 add from set2: ",set1)
# add letters by add()
set2.add("t")
set2.add("u")
print("add \"t\" \"u\" to set2: ",set2)
```
